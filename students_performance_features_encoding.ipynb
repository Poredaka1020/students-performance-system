```python
# Fonction pour modifier la colonne "Ethnicity" en string
# Ethnicity strings values {0: Caucasian, 1: African American, 2: Asian, 3: Other}
# def update_ethnicity_to_string(column : int) -> str:
#     """Function to update the integer values to string

#     Args:
#         column: the column that contains all the values they are updated to string

#     Returns:
#         str: the all values updated
#     """
#     if column == 0:
#         return "Caucasian"
#     elif column == 1:
#         return "African_American"
#     elif column == 2:
#         return "Asian"
#     return "Other"

# ---------------
# Fonction pour modifier la colonne "GradeClass" en string
def update_grade_class_to_string(column : float) -> str:
    """Function to update the GradeClass to the string characters

    Args:
        column: the column that contains all the values they are updated to string

    Return:
        str: the all values updated
    """
    if column >= 3.5:
        return "A"
    elif (column >= 3.0) & (column < 3.5):
        return "B"
    elif (column >= 2.5) & (column < 3.0):
        return "C"
    elif (column >= 2.0) & (column < 2.5):
        return "D"
    return "F"
```

```python
# Un dictionnaire pour chaque colonne
gender_dict = {0: "Male", 1: "Female"}
ethnicity_dict = {0: "Caucasian", 1: "African_American", 2: "Asian", 3: "Other"}
parental_educ_dict = {0: None, 1: "High_School", 2: "Some_College", 3: "Bachelor's", 4: "Higer"}
tutoring_dict = {0: "No", 1: "Yes"}
parental_support_dict = {0: None, 1: "Low", 2: "Moderate", 3: "High", 4: "Very_High"}
extra_curricular_dict = {0: "No", 1: "Yes"}
sports_dict = {0: "No", 1: "Yes"}
music_dict = {0: "No", 1: "Yes"}
volunteering_dict = {0: "No", 1: "Yes"}
# Première approche pour modifier les colonnes. Ex: 
# df_students_perf["Ethnicity"] = df_students_perf["Ethnicity"].replace(ethnicity_dict)
# Deuxième approche avec une fonction 
# _studdf_students_perf["Ethnicity"] = df_students_perf["Ethnicity"].apply(update_ethnicity_to_string)
```

```python
# Remplacer directement les valeurs de la colonne cible dans le dataframe. Ecraser les anciennes valeurs des colonnes
# Pour Ethnicity
df_students_perf.replace({"Ethnicity" : ethnicity_dict}, inplace=True)
# Pour Gender
df_students_perf.replace({"Gender" : gender_dict}, inplace=True)
# Pour ParentalEducation
df_students_perf.replace({"ParentalEducation" : parental_educ_dict}, inplace=True)
# Pour Tutoring
df_students_perf.replace({"Tutoring" : tutoring_dict}, inplace=True)
# Pour ParentalSupport
df_students_perf.replace({"ParentalSupport" : parental_support_dict}, inplace=True)
# Pour Extracurricular
df_students_perf.replace({"Extracurricular" : extra_curricular_dict}, inplace=True)
# Pour Sports
df_students_perf.replace({"Sports" : sports_dict}, inplace=True)
# Pour Music
df_students_perf.replace({"Music" : music_dict}, inplace=True)
# Pour Volunteering
df_students_perf.replace({"Volunteering" : volunteering_dict}, inplace=True)
```

```python
df_students_perf["GradeClass"] = df_students_perf["GPA"].apply(update_grade_class_to_string)
df_students_perf
```
